```python
import pandas as pd
import numpy as py
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import sklearn
print(sklearn.__version__)
```

```python
from sklearn.datasets import load_iris
data = load_iris()
#print((data['data']))
df = pd.DataFrame(data['data'],columns=['sepal length','sepal width','petal length','petal width'])
```

```python
df
```

```python
print((data))
df2 = pd.DataFrame(data['target'],columns=['class'])
df2
```

```python
master_df = pd.concat([df,df2],axis=1)
master_df
```

```python
master_df['class'] = master_df['class'].map({0:'setosa',1:'versicolor',2:'virginica'})
```

```python
master_df
```

```python
sns.lmplot(data=master_df,x='sepal length',y='sepal width',hue='class',fit_reg=False)
sns.lmplot(data=master_df,x='petal length',y='petal width',hue='class',fit_reg=False)
```

```python
master_df
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
master_df['class']=label_encoder.fit_transform(master_df['class'])
```

```python
from sklearn.model_selection import train_test_split
X = master_df.iloc[: , :4]
Y = master_df['class']
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.3,random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
Decision_Tree_Classifier = DecisionTreeClassifier(min_samples_split=2)
Decision_Tree_Classifier.fit(x_train,y_train)
```

```python
y_pred = Decision_Tree_Classifier.predict(x_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.naive_bayes import GaussianNB
NB = GaussianNB()
NB.fit(x_train,y_train)
```

```python
y_pred = NB.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
from sklearn.neighbors import KNeighborsClassifier
k_neighbors_classifier = KNeighborsClassifier()
k_neighbors_classifier.fit(x_train,y_train)
```

```python
y_pred = k_neighbors_classifier.predict(x_test)
accuracy_score(y_test,y_pred)
```
